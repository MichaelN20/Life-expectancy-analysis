### **5.2. What Factors Can Affect Someone's Life Expectancy?**

To answer this question, we will perform a `chi-squared test` because we want to test the `relationship between two factors`.

To conduct the test, we need to formulate hypotheses.

`H0: There is no relationship between the variables and Life expectancy.`

`H1: There is a relationship between the variables and Life expectancy.`

The confidence level used is 95%.

```python
# Creating a new DataFrame without the Status and Life expectancy columns.
'''
The reason for not using these columns is that we do not want to analyze the relationship between Status,
the number of adult deaths, the number of infant deaths, the number of under-five deaths, and Life expectancy
with those columns themselves.
'''
cols = data_for_statistic.drop(columns=['Status', 'Adult Mortality', 'infant deaths', 'Life expectancy', 'under-five deaths', 'Year']).columns

# Creating a loop containing the chi-squared test
for col in cols:
  # Chi-square formula
  cross = pd.crosstab(data_for_statistic[col], data_for_statistic['Life expectancy'])
  pval = stats.chi2_contingency(cross).pvalue

  # Creating an if statement to display loop results based on the p-value
  if pval == 0:
    print(f"P-value from {col} and Life expectancy: {pval}, invalid!")
  elif pval < 0.05:
    print(f"P-value from {col} and Life expectancy: {pval}, they are related")
  else:
    print(f"P-value from {col} and Life expectancy: {pval}")
```

Based on the obtained results, the factors related to life expectancy are:
- The country where they live `(Country)`
- Hepatitis B immunization `(Hepatitis B)`
- Average Body Mass Index of the entire population `(BMI)`
- Polio immunization `(Polio)`
- Government general health expenditure as a percentage of total government expenditure (%) `(Total expenditure)`
- Prevalence of stunting in children and adolescents aged 10 to 19 years (%) `(thinness 1-19 years)`
- Prevalence of stunting in children aged 5 to 9 years (%) `(thinness 5-9 years)`
- Number of years of schooling (years) `(Schooling)`

### **5.3. Does Life Expectancy have a positive or negative correlation with factors unrelated to diseases?**

To answer this question, we will perform `correlation tests (Pearson, Spearman, and Kendall)`.

However, before determining the type of correlation test, we need to assess the data distribution using the `skewness test`.

The skewness test is used to `measure how skewed or skewed the data distribution is from a normal distribution`. Skewness is a measure of the asymmetry of the data distribution. If the skewness is positive, the tail of the distribution will be skewed to the right. Conversely, if negative, the tail of the distribution will be skewed to the left.

- If Skewness = 0, indicates a `symmetric tail distribution.`
- If Skewness > 0, indicates a `right-skewed tail distribution (positive skewness).`
- If Skewness < 0, indicates a `left-skewed tail distribution (negative skewness).`

#### **5.3.1. Determining the Type of Data Distribution with Skewness Test**

```python
# Create a DataFrame 'cols' to store the columns to be used in the skewness test
# The 'Country' column does not need skewness testing because it is an object/string
cols = ['BMI', 'Total expenditure', 'Schooling']

# Create an empty dictionary to store column names and their skewness values
skewness_dict = {}

# Iterate through each column in the DataFrame
for column in cols:
    # Calculate skewness for the current column
    skewness_value = data_for_statistic[column].skew()
    
    # Store the column name and its skewness value in the dictionary
    skewness_dict[column] = skewness_value

# Convert the dictionary into a pandas Series for easier viewing
skewness_series = pd.Series(skewness_dict)

# Display the skewness values for each column
print(skewness_series)
```

From the skewness test results, we find that the columns `BMI, Total expenditure, Schooling` have data that is normally or symmetrically distributed.

This is seen in the skewness values `between -0.5 and 0.5`, indicating a symmetrical distribution.

Therefore, we will use the `pearson` correlation test for data with a normal distribution.

The Pearson correlation test is used to `measure the extent of the linear relationship between two variables`. The result is the `Pearson correlation coefficient (r)`, which can range from -1 to 1.

- If r = 1, indicates a `perfect positive linear relationship.`
- If r = -1, indicates a `perfect negative linear relationship.`
- If r = 0, indicates `no linear relationship.`

#### **5.3.2. Does Life Expectancy have a positive or negative correlation with BMI?**

To perform the Pearson correlation test for the correlation between BMI and Life expectancy, we need to formulate hypotheses.

`H0: BMI HAS A CORRELATION with Life expectancy.`

`H1: BMI DOES NOT HAVE A CORRELATION with Life expectancy.`

```python
# Using the Pearson correlation test formula to obtain r-correlation
corr_r, pval_p = stats.pearsonr(data_for_statistic['BMI'], data_for_statistic['Life expectancy'])
print(f"r-correlation: {corr_r:.2f}, p-value: {pval_p}")
```

```python
# Create a scatter plot with a trend line
sns.regplot(x='BMI', y='Life expectancy', data=data_for_statistic, line_kws={'color': 'blue'})

# Add title and labels
plt.title(f'Correlation between BMI and Life expectancy (Pearson Correlation: {corr_r:.2f}, p-value: {pval_p:.4f})')
plt.xlabel('BMI')
plt.ylabel('Life Expectancy')

# Display the plot
plt.show()
```

From the Pearson correlation test results, we obtained a correlation coefficient of `0.54`, indicating that BMI has a positive correlation with Life Expectancy.

A positive correlation means that as the `average Body Mass Index of the entire population (BMI)` increases, the `Life Expectancy` also increases.

From the visualization results, we can see that someone's BMI has a quite significant impact on life expectancy. We can also observe that a life expectancy above 65 years is achieved by those with a BMI in the range of 45 to 55.

### **5.4. How does the average life expectancy compare between the top 5 and bottom 5 countries in 2015?**

To compare the `countries with the highest life expectancy` against the `countries with the lowest life expectancy` in 2015, we can display it with a `bar chart where each category displays 5 countries`.

```python
# Filter data only for the Year 2015
data_2015 = data_for_visualization_no_statistic[data_for_visualization_no_statistic['Year'] == 2015]

# Calculate the average Life Expectancy for each country
avg_life_expectancy_by_country = data_2015.groupby('Country')['Life expectancy'].mean().reset_index()

# Sort by average Life Expectancy in descending order
avg_life_expectancy_by_country = avg_life_expectancy_by_country.sort_values(by='Life expectancy', ascending=False)

# Take the top 5 countries with the highest averages and the bottom 5 countries with the lowest averages
top_5_countries = avg_life_expectancy_by_country.head(5)
bottom_5_countries = avg_life_expectancy_by_country.tail(5)

# Combine both for visualization
combined_df = pd.concat([top_5_countries, bottom_5_countries])

# Plotting the visualization
sns.barplot(x='Life expectancy', y='Country', data=combined_df, palette='crest')

# Add title and labels
plt.title('Top 5 and Bottom 5 Average Life Expectancy of Each Country (2015)')
plt.xlabel('Average Life Expectancy')
plt.ylabel('Country')

# Display the plot
plt.show()
```

From the visualization results, we can conclude that the difference between the highest and lowest life expectancies is around 30 years. Looking at the results, the `bottom 5 countries are countries classified as poor countries`. Poor countries generally `face challenges in education, nutrition, and government health assistance`, which usually occurs due to `financial limitations of a country`. This reinforces the earlier analysis results on the correlation of BMI, education level, and government health assistance, making their correlations more strongly supported.
